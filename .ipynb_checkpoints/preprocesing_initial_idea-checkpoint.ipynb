```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("C:\Users\jeanl\College\Blocks\Block 3\Epic\Block3Epic\archive\Breast_Cancer.csv")
data.head()

X = data[["Age"]].values
y = data[["Tumor Size"]].values
```

```python
# Visualize the age vs tumor size actual
import matplotlib as plt
data.plot(kind='scatter', grid=True,
             x="Age", y="Tumor Size")
# plt.axis([23_500, 62_500, 4, 9])
# plt.show()
```

```python
import matplotlib.pyplot as plt
#Visualise age vs average tumour size per age( before removing outliers)
average_tumor_size = data.groupby('Age')['Tumor Size'].mean()
plt.scatter(average_tumor_size.index, average_tumor_size.values)
```

```python
#removing outliers based on age/tumour size
#I'm doing this by finding the interquartile range (where most values are), and getting rid of any values outside of that.
# Calculate the interquartile range (IQR)
Q1 = data["Tumor Size"].quantile(0.25)
Q3 = data["Tumor Size"].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


# Identify and remove outliers
outliers = (data["Tumor Size"] < lower_bound) | (data["Tumor Size"] > upper_bound)
tumor_size_without_outliers = data[~outliers]
average_tumor_size_no_outliers = tumor_size_without_outliers.groupby('Age')['Tumor Size'].mean()

plt.scatter(average_tumor_size_no_outliers.index, average_tumor_size_no_outliers.values, label = "No outliers", color = 'red')
plt.scatter(average_tumor_size.index, average_tumor_size.values, label = "with outliers", color = "blue")

plt.title('Average Tumor Size for Each Age')
plt.xlabel('Age')
plt.ylabel('Average Tumor Size')
plt.legend()
plt.show()
```

```python
#Doing the same as above, but removing outliers based on age, not overall outliers
#removing outliers based on age/tumour size
#I'm doing this by finding the interquartile range (where most values are), and getting rid of any values outside of that.
# Calculate the interquartile range (IQR)
final_no_outliers_by_age = {}
for age, age_data in data.groupby("Age"):
    Q1 = age_data["Tumor Size"].quantile(0.25)
    Q3 = age_data["Tumor Size"].quantile(0.75)
    IQR = Q3 - Q1

    # Define the lower and upper bounds to identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    # Identify and remove outliers
    outliers = (age_data["Tumor Size"] < lower_bound) | (age_data["Tumor Size"] > upper_bound)
    tumor_size_without_outliers = age_data[~outliers]
    average_tumor_size_no_outliers = tumor_size_without_outliers["Tumor Size"].mean()
    final_no_outliers_by_age.update({age: average_tumor_size_no_outliers})

plt.scatter(final_no_outliers_by_age.keys(), final_no_outliers_by_age.values(), label = "No outliers", color = 'red')
plt.scatter(average_tumor_size.index, average_tumor_size.values, label = "with outliers", color = "blue")

plt.title('Average Tumor Size for Each Age')
plt.xlabel('Age')
plt.ylabel('Average Tumor Size')
plt.legend()
plt.show()
```
