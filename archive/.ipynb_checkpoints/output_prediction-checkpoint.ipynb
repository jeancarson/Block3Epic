```python
import pandas as pd
df = pd.read_csv('Breast_Cancer.csv')
race_dummie = pd.get_dummies(df.Race).astype(int)
marriage_dummie = pd.get_dummies(df['Marital Status']).astype(int)
t_dummie = pd.get_dummies(df['T Stage ']).astype(int)
n_dummie = pd.get_dummies(df['N Stage']).astype(int)
six_dummie = pd.get_dummies(df['6th Stage']).astype(int)
diff_dummie = pd.get_dummies(df.differentiate).astype(int)
a_dummie = pd.get_dummies(df['A Stage']).astype(int)
e_dummie = pd.get_dummies(df['Estrogen Status']).astype(int)
p_dummie = pd.get_dummies(df['Progesterone Status']).astype(int)
df
```

```python
merged = pd.concat([df,race_dummie,marriage_dummie,t_dummie,n_dummie,six_dummie,diff_dummie,a_dummie,e_dummie,p_dummie],axis='columns')
```

```python
final = merged.drop(['Race', 'Marital Status', 'T Stage ', 'N Stage', '6th Stage', 'differentiate', 'A Stage', 'Estrogen Status', 'Progesterone Status'])
```
