```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
df = pd.read_csv('/Users/miaborko/Documents/epic3/Block3Epic/cleaned_data.csv')  #here is where visuals are happening but not yet
```

```python
print(df.columns)
df.head()
```

```python

fig, ax = plt.subplots(figsize=(8,5), dpi=100)
sns.countplot(data=df, x='N Stage', palette='mako_r', ax=ax)
ax.set_xlabel("Status")
ax.set_ylabel("Number of patients")
ax.set_title("Status of N stage ")
percentage = round(df['N Stage'].value_counts() * 100 /len(df), 2)
patches = ax.patches
for i in range(len(patches)):
   x = patches[i].get_x() + patches[i].get_width()/2
   y = patches[i].get_height()+.05
   ax.annotate('{:.2f}%'.format(percentage[i]), (x, y), ha='center')

plt.show()
```

```python
box_cols = ['Survival Months', 'Tumor Size (mm)', 'Regional Node Positive']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(14,4), dpi=120)
fig.suptitle('Distribution of Continuous Features', y = 0.98);
for col, ax in zip(box_cols, axes.ravel()):
    sns.boxplot(data=df, x='Status', y=col, palette='mako_r', ax=ax)
plt.tight_layout()
```
